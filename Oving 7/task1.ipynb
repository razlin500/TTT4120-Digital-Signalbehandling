```python
import numpy as np
import matplotlib.pyplot as plt
import random as r
```

```python
rng = np.random.default_rng()
```

```python
#Task 1A: Generate and plot three types of white noise: binary, Gaussian, and uniform.

bool_array = [rng.choice([-1, 1]) for i in range(100)]
normal_array = [rng.standard_normal() for i in range(100)]
unif_array = [rng.uniform(-1, 1) for i in range(100)]

fig, axes = plt.subplots(3, 1, figsize=(10, 8), constrained_layout=True)

axes[0].stem(bool_array, linefmt='b-', markerfmt='bo', basefmt='r-', label='White binary noise')
axes[0].set_title('White Binary Noise')
axes[0].set_xlabel('Sample Index')
axes[0].set_ylabel('Amplitude')
axes[0].grid(True)
axes[0].legend().remove()

axes[1].stem(normal_array, linefmt='g-', markerfmt='go', basefmt='r-', label='White Gaussian noise')
axes[1].set_title('White Gaussian Noise')
axes[1].set_xlabel('Sample Index')
axes[1].set_ylabel('Amplitude')
axes[1].grid(True)
axes[1].legend().remove()

axes[2].stem(unif_array, linefmt='m-', markerfmt='mo', basefmt='r-', label='White uniform noise')
axes[2].set_title('White Uniform Noise')
axes[2].set_xlabel('Sample Index')
axes[2].set_ylabel('Amplitude')
axes[2].grid(True)
axes[2].legend().remove()

plt.suptitle('Comparison of Different Types of White Noise', fontsize=16)
plt.show()
```

```python
#Task 1C: 

import scipy.signal as signal



long_normal = [rng.standard_normal() for i in range(20000)]
long_uniform = [rng.uniform(-1, 1) for i in range(20000)]
long_bool = [rng.choice([-1, 1]) for i in range(20000)]

normal_mean = np.mean(long_normal)
uniform_mean = np.mean(long_uniform)
bool_mean = np.mean(long_bool)

print(f"Mean of long Gaussian noise: {normal_mean}")
print(f"Mean of long Uniform noise: {uniform_mean}")
print(f"Mean of long Binary noise: {bool_mean}")


corr_normal = signal.correlate(long_normal, long_normal, mode='full')
corr_normal = corr_normal[::2857]
corr_uniform = signal.correlate(long_uniform, long_uniform, mode='full')
corr_uniform = corr_uniform[::2857]
corr_bool = signal.correlate(long_bool, long_bool, mode='full')
corr_bool = corr_bool[::2857]   

fig, axes = plt.subplots(3, 1, figsize=(10, 8), constrained_layout=True)

axes[0].stem(corr_bool, label='Autocorrelation of Binary noise')
axes[0].set_title('Autocorrelation of White Binary Noise')
axes[0].set_xlabel('Lag')
axes[0].set_ylabel('Autocorrelation')

axes[1].stem(corr_normal, label='Autocorrelation of Gaussian noise')
axes[1].set_title('Autocorrelation of White Gaussian Noise')
axes[1].set_xlabel('Lag')
axes[1].set_ylabel('Autocorrelation')

axes[2].stem(corr_uniform, label='Autocorrelation of Uniform noise')
axes[2].set_title('Autocorrelation of White Uniform Noise')
axes[2].set_xlabel('Lag')
axes[2].set_ylabel('Autocorrelation')
```
