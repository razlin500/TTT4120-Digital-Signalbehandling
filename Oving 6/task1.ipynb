```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Task 1A

def X(f):
    return (1-(0.9*np.exp(-1j*2*np.pi*f))**28)/(1-0.9*np.exp(-1j*2*np.pi*f))

f = np.linspace(0, 1, 1000)

analytical_X = np.abs(X(f))

plt.figure(figsize=(10, 10))
plt.plot(f, analytical_X, linewidth=2, color='#2E86C1', label='|X(f)|')
plt.title('Magnitude Spectrum |X(f)|', fontsize=14, pad=15)
plt.xlabel('Frequency (f)', fontsize=12)
plt.ylabel('Magnitude', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=10)
plt.tight_layout()
```

```python
#Task 1B

N_x = 28
x_n = np.array([0.9**n for n in range(N_x)])

# Calculate FFTs with different N values
X_1 = np.fft.fft(x_n, int(N_x*0.25))  # N = 7
X_2 = np.fft.fft(x_n, int(N_x*0.5))   # N = 14
X_3 = np.fft.fft(x_n, N_x)            # N = 28
X_4 = np.fft.fft(x_n, N_x*2)          # N = 56

```

```python
#Task 1C

# For demonstration, I'll include minimal computation
# Replace this with your actual variables from tasks (a), (b), (c)
Nx = 28
n = np.arange(Nx)
x = 0.9**n

# Your DTFT from task (a)
f_dtft = f
X_dtft = analytical_X

# Your DFTs from task (b)
X_dft_7 = X_1
X_dft_14 = X_2
X_dft_28 = X_3
X_dft_56 = X_4
# Your frequency values from task (c): f = k/N
f_dft_7 = np.arange(7) / 7
f_dft_14 = np.arange(14) / 14
f_dft_28 = np.arange(28) / 28
f_dft_56 = np.arange(56) / 56

# Task 1d: Create plots
fig, axes = plt.subplots(2, 2, figsize=(12, 12))
axes = axes.flatten()

# Pastel colors
dtft_color = '#6B9BD1'  # Soft blue
dft_color = '#E57373'   # Soft red/coral

# Plot 1: N = Nx/4 = 7
axes[0].plot(f_dtft, np.abs(X_dtft), color=dtft_color, linewidth=2, label='DTFT')
markerline, stemlines, baseline = axes[0].stem(f_dft_7, np.abs(X_dft_7), linefmt=dft_color, markerfmt='o', label='DFT')
baseline.set_visible(False)
markerline.set_markerfacecolor(dft_color)
markerline.set_markeredgecolor(dft_color)
markerline.set_markersize(6)
stemlines.set_linewidth(1.5)
axes[0].set_xlabel('Normalized Frequency f')
axes[0].set_ylabel('Magnitude')
axes[0].set_title('DFT (N=7) vs DTFT')
axes[0].legend()
axes[0].grid(True, alpha=0.3)
axes[0].set_xlim([0, 1])

# Plot 2: N = Nx/2 = 14
axes[1].plot(f_dtft, np.abs(X_dtft), color=dtft_color, linewidth=2, label='DTFT')
markerline, stemlines, baseline = axes[1].stem(f_dft_14, np.abs(X_dft_14), linefmt=dft_color, markerfmt='o', label='DFT')
baseline.set_visible(False)
markerline.set_markerfacecolor(dft_color)
markerline.set_markeredgecolor(dft_color)
markerline.set_markersize(6)
stemlines.set_linewidth(1.5)
axes[1].set_xlabel('Normalized Frequency f')
axes[1].set_ylabel('Magnitude')
axes[1].set_title('DFT (N=14) vs DTFT')
axes[1].legend()
axes[1].grid(True, alpha=0.3)
axes[1].set_xlim([0, 1])

# Plot 3: N = Nx = 28
axes[2].plot(f_dtft, np.abs(X_dtft), color=dtft_color, linewidth=2, label='DTFT')
markerline, stemlines, baseline = axes[2].stem(f_dft_28, np.abs(X_dft_28), linefmt=dft_color, markerfmt='o', label='DFT')
baseline.set_visible(False)
markerline.set_markerfacecolor(dft_color)
markerline.set_markeredgecolor(dft_color)
markerline.set_markersize(6)
stemlines.set_linewidth(1.5)
axes[2].set_xlabel('Normalized Frequency f')
axes[2].set_ylabel('Magnitude')
axes[2].set_title('DFT (N=28) vs DTFT')
axes[2].legend()
axes[2].grid(True, alpha=0.3)
axes[2].set_xlim([0, 1])

# Plot 4: N = 2*Nx = 56
axes[3].plot(f_dtft, np.abs(X_dtft), color=dtft_color, linewidth=2, label='DTFT')
markerline, stemlines, baseline = axes[3].stem(f_dft_56, np.abs(X_dft_56), linefmt=dft_color, markerfmt='o', label='DFT')
baseline.set_visible(False)
markerline.set_markerfacecolor(dft_color)
markerline.set_markeredgecolor(dft_color)
markerline.set_markersize(6)
stemlines.set_linewidth(1.5)
axes[3].set_xlabel('Normalized Frequency f')
axes[3].set_ylabel('Magnitude')
axes[3].set_title('DFT (N=56) vs DTFT')
axes[3].legend()
axes[3].grid(True, alpha=0.3)
axes[3].set_xlim([0, 1])

fig.suptitle('Comparison of DFT and DTFT Magnitudes', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```
