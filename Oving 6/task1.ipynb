```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Task 1A

def X(f):
    return (1-(0.9*np.exp(-1j*2*np.pi*f))**28)/(1-0.9*np.exp(-1j*2*np.pi*f))

f = np.linspace(0, 1, 1000)

plt.figure(figsize=(10, 10))
plt.plot(f, np.abs(X(f)), linewidth=2, color='#2E86C1', label='|X(f)|')
plt.title('Magnitude Spectrum |X(f)|', fontsize=14, pad=15)
plt.xlabel('Frequency (f)', fontsize=12)
plt.ylabel('Magnitude', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=10)
plt.tight_layout()
```
