```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
a_values = [0.5, 0.9, -0.9]
n = np.arange(0, 51)
l = np.arange(-50, 51)
f = np.linspace(-0.5, 0.5, 1000)

plt.figure(figsize=(12, 9))

for row, a in enumerate(a_values):
    # Compute signals
    x = a ** n
    r_xx = a ** np.abs(l) / (1 - a ** 2)
    S_xx = 1 / (1 + a ** 2 - 2 * a * np.cos(2 * np.pi * f))
    
    # Column 1: x[n]
    plt.subplot(3, 3, 3 * row + 1)
    plt.stem(n, x, linefmt='-', markerfmt='.', basefmt=' ')
    plt.title(f"x[n], a = {a}")
    plt.xlabel("n")
    plt.ylabel("x[n]")
    
    # Column 2: r_xx(l)
    plt.subplot(3, 3, 3 * row + 2)
    plt.stem(l, r_xx, linefmt='-', markerfmt='.', basefmt=' ')
    plt.title(f"r_xx(l), a = {a}")
    plt.xlabel("l")
    plt.ylabel("r_xx(l)")
    
    # Column 3: S_xx(f)
    plt.subplot(3, 3, 3 * row + 3)
    plt.plot(f, S_xx)
    plt.title(f"S_xx(f), a = {a}")
    plt.xlabel("f")
    plt.ylabel("S_xx(f)")

plt.tight_layout()
plt.show()
```
